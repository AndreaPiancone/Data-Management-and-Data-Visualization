```python
from pyArango.connection import *
import pandas as pd
```

```python
conn = Connection(username = "root", password = "arango")
```

```python
db = conn['GamesDB']
```

### Query visualizzazioni medie per genere

```python
sub_query = "LET S = (FOR game IN games FOR i IN game.genres FOR k in game.views RETURN({'name': i.name, 'visuals':k.visuals}))"
query = "FOR i in S COLLECT name = i.name INTO g RETURN({name: name, views_medie: AVG(g[*].i.visuals), sd: STDDEV_SAMPLE(g[*].i.visuals), n:LENGTH(g), l_bound: AVG(g[*].i.visuals)-1.96*(STDDEV_SAMPLE(g[*].i.visuals) / SQRT(LENGTH(g))), u_bound: AVG(g[*].i.visuals)+1.96*(STDDEV_SAMPLE(g[*].i.visuals) / SQRT(LENGTH(g)))})"
```

```python
Query_genere = sub_query + " " + query
```

```python
queryResult = db.AQLQuery(Query_genere, rawResults=True, batchSize=100)
```

```python
genere = pd.DataFrame(queryResult)
```

```python
genere
```

```python
genere.to_csv(r"/home/andrea/Scrivania/genere.csv", index = False)
```
