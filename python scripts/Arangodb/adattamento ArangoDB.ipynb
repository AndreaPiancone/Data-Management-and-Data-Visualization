```python
import pandas as pd
import json
```

```python
games = []
for i in open(r"C:\Users\39331\Desktop\Integrato_cleaned.json"):
    games.append(json.loads(i))
```

```python
document = []
for i in games:
    document.append({'name':i['name'], 
                    'genres':i['genres'],
                    'themes':i['themes'],
                    'characters':i['characters'],
                    'dlcs':i['dlcs'],
                    'game_modes':i['game_modes'],
                    'views':i['views'],
                    'age_rating':i['age_rating'],
                    'original_release_date':i['original_release_date'],
                     'anno_pubblicazione': i['anno_pubblicazione'],
                    'platforms':i['platforms']}) #nodi del grafo
```

```python
edges = []
for i in games:
    nome = i['name']
    try:
        for j in i['similar_games']:
            edges.append({'_from':"games/" + nome,
                     "_to":"games/"+j['name']})
    except:
        pass
```

```python
with open(r'C:\Users\39331\Desktop\edges.json', 'w') as outfile:
    json.dump(edges, outfile)
```

```python
with open(r'C:\Users\39331\Desktop\documents.json', 'w') as outfile:
    json.dump(document, outfile)
```
